```python
#This function can fit you with a multi linear regression model(up to 5 variables)
#All you need to do is choose target column and variable columns and then it will give you the answer 
#based on R square if it is larger than 0.6. 

def lm_decider():
    #import necessary libraries and pakages
    import pandas as pd
    from pandas import DataFrame,Series
    import numpy as np
    from sklearn import linear_model
    import os
    
    #set default address for further convenience
    os.getcwd() 
    os.chdir("C:\\Users\\yiyao\\Desktop\\6983")
    df = pd.read_csv("data.csv")
    
    #check dataset
    print(df.head(5))
    col = df.columns.tolist()#store column names in list
    
    #let user input which column they want to use 
    number = input("Enter column number you want use as target column,start from 0 :")
    number = int(number)
    Max_number = len(col) + 1
    if number > Max_number:
        number = input("Re-enter a valid number,start from 0 :")
    number = int(number)
    target = df[[col[number]]]
    print(target.head(5))
    
    #following function is inspired by Sri Bandhakavi's function
    var = []
    MaxlenVar = 5
    while len(var) < MaxlenVar:
        var_num = input("Enter column numbers you want to use as variable columns up to 5 (or nothing to skip): ")
        if var_num == '':
            break
        var_num=int(var_num)
        var.append(var_num)
    print(str(var) + " <- Check column numbers entered. If incorrect, skip next step and restart") # allows user to check if entries are correct
    print("")
    
    #Putting other choosen columns into variables dataset
    variable = df[[col[var[0]]]]
    length = len(variable)
    name = df.columns
    for i in range(len(var)):
        variable[name[var[i]]] = Series(df[name[var[i]]], index=variable.index)
        i = i+1
    variable.head(5)
    x = variable
    y = target
    lm = linear_model.LinearRegression()
    model = lm.fit(x,y)
    if lm.score(x,y) < 0.7:
        print("The linear regression model is BAD as R-square is {}.".format(lm.score(x,y)))
    else:
        print("The linear regression model is GOOD as R-square is {}.".format(lm.score(x,y)))
lm_decider()
```
